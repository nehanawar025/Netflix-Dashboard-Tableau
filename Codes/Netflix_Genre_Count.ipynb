```python
import pandas as pd
```

```python
df = pd.read_csv('/content/genre.csv')
```

```python
df
```

```python
df['genre_list'] = df['listed_in'].str.split(', ')

df
```

```python
# Flatten the list of lists into a single list
all_genres = [genre for sublist in df['genre_list'] for genre in sublist]

all_genres
```

```python
# Get unique genres
unique_genres = list(set(all_genres))

print(unique_genres)
```

```python
from collections import Counter

# Count the occurrences of each genre
genre_counts = Counter(all_genres)
```

```python
genre_counts
```

```python
# Create a list of tuples (genre, count)
genre_counts_list = list(genre_counts.items())
```

```python
# Sort the list by count in descending order
genre_counts_list.sort(key=lambda x: x[1], reverse=True)
```

```python
genre_counts_list
```

```python
# Create a DataFrame from the sorted list
output_df = pd.DataFrame(genre_counts_list, columns=['genre', 'count'])
```

```python
output_df.to_csv('genre_counts.csv', index=False)
```
